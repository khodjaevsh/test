```python
import featuretools as ft
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv('search logs processed.csv', parse_dates=['departure','created_date'], na_values = "NaN")

data.head()
```

```python
data.info()
```

```python
data.head()
```

# Drop data

```python
data.drop(['trace_id', 'created_at','client_ruid','client_ip'], axis =1, inplace = True)
```

## Treat user_agent

```python
data['user_agent'] = data['user_agent'].astype('str')
data['user_agent'] = [x.split('/')[0] for x in data['user_agent']]
```

```python
data.head()
```

## IO columns

```python
def io(df,columns):
    for col in columns:
        df[col] = df[col].astype('str')
        df[col] = [0 if len(x) < 4 else 1 for x in df[col]]
        
io(data,['install_id', 'return', 'webtrekk_eid'])
```

## True/False into IO

```python
def tf(df,columns):
    for col in columns:
        df[col] = df[col].astype('str')
        df[col] = [1 if len(x) < 5 else 0 for x in df[col]]
        
tf(data,['after_midnight_rides', 'back'])
```

```python
data.head()
```

```python
es  =  ft.EntitySet("sl")
```

```python
from featuretools import variable_types as vtypes

trip_variable_types = {
    'after_midnight_rides': vtypes.Categorical,
    'back': vtypes.Categorical,
    'bikes': vtypes.Categorical,
    'children': vtypes.Ordinal,
    'currency': vtypes.Categorical,
    'from': vtypes.Categorical,
    'return': vtypes.Categorical,
    'search_by': vtypes.Categorical,
    'to': vtypes.Categorical,
    'user_agent': vtypes.Categorical,
    'bikes': vtypes.Categorical,
    'install_id': vtypes.Categorical,
    'webtrekk_eid': vtypes.Categorical,
    'shop_code': vtypes.Categorical
}

es.entity_from_dataframe(entity_id="sl",
                         dataframe=data,
                         index="id",
                         variable_types=trip_variable_types)
```

```python
es.normalize_entity(base_entity_id="sl",
                    new_entity_id="from1",
                    index="from")

es.normalize_entity(base_entity_id="sl",
                    new_entity_id="to1",
                    index="to")

es.normalize_entity(base_entity_id="sl",
                    new_entity_id="WT",
                    index="webtrekk_eid")
```

```python
agg_primitives = ['Sum', 'Mean', 'Median', 'Std', 'Count', 'Min', 'Max', 'Skew']
trans_primitives = ['Day', 'Hour', 'Minute', 'Month', 'Weekday', 'Week', 'Weekend']


es.add_interesting_values(100)

feature_matrix, features = ft.dfs(entityset=es,
                                  target_entity="sl",
                                  trans_primitives=trans_primitives,
                                  agg_primitives=agg_primitives,
                                  verbose=True,
                                  max_depth=5)
feature_matrix.head()
```
